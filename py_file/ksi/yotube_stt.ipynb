```python
import os
import whisper

ffmpeg_path = "C:\\prg\\ffmpeg\\ffmpeg-master-latest-win64-gpl-shared\\bin"

AUDIO_DIR = "./data/Audio"  # 오디오 추출 저장 딕렉토리
TEXT_DIR = "./data/Text"    # 자막 저장 디렉토리

# 디렉토리 확인후 없으면 생성성
os.makedirs(TEXT_DIR, exist_ok=True)

# ffmpeg 경로 명시
os.environ["PATH"] += os.pathsep + rf"{ffmpeg_path}"

# whisper 모델 설정
model = whisper.load_model("small").to("cuda")

for file in os.listdir(AUDIO_DIR):
    if file.endswith(".mp3"):
        audio_path = os.path.join(AUDIO_DIR, file)
        output = os.path.join(TEXT_DIR, file.replace(".mp3",".txt"))
        
        # 중복 여부 확인
        if os.path.exists(output):
            print(f"이미 존재함, 건너뜀: {file}")
            continue

        print(f"처리 시작: {file}")

        try:
            # 오디오 -> 텍스트 처리
            result = model.transcribe(audio_path, language="ko", fp16=True)
            text = result["text"]

            
            # 자막 TEXT 저장
            with open(output, 'w', encoding='utf-8') as f:
                f.write(text)

            print(f"처리 완료: {file}")
        
        except Exception as e:
            print(f"에러 : {file}")
```

```python
from llm.call_llm import call_llm
import json
import re

def create_json_with_llm(dog_text):
    system_prompt=""
    user_prompt=""

    system_prompt = f'''
    당신은 반려견 전문가입니다. 사용자가 입력한 견종 설명 텍스트를 기반으로 다음 JSON 구조를 생성하세요:

    1. text_chunks  = 의미 단위로 300~500자 이내로 분할
    2. id = breed_(견종 영어 이름)
    3. title = (경종 한국 이름) 견종 정보
    4. metadata의 결과값은 영어로만 출력
    5. 오타같거나 문제가 되는 텍스트는 행당 문맥에 맞게 수정
    5. 다음 형식을 JSON으로만 정확히 출력하세요. 설명 금지
    반환 형식:
    {{
      "id": "...",
      "title": "...",
      "text_chunks": [...],
      "metadata": {{
        "origin": "...",
        "size": "...",
        "breed_group": "...",
        "fur_type": "...",
        "trainability": "...",
        "activity_level": "...",
        "separation_anxiety": true,
        "family_friendly": true,
        "aggression_tendency": "...",
        "health_issues": [...],
        "grooming_needs": "...",
        "guard_ability": "...",
        "difficulty_score": ...
      }}
    }}
    '''

    user_prompt = f'''
    아래는 해당 견종에 대한 설명입니다:
    {dog_text}
    '''

    response = call_llm(system_prompt, user_prompt)
    
    try:
        return json.loads(response)
    except json.JSONDecodeError:
        match = re.search(r'\{.*\}', response, re.DOTALL)
        if match:
            return json.loads(match.group())
        else:
            raise ValueError("GPT 응답에서 JSON을 추출하지 못했습니다.")
```

```python
import os
import json

TEXT_DIR = "./data/Text"
JSON_DIR = "./data/Json"
os.makedirs(JSON_DIR, exist_ok=True)

for file in os.listdir(TEXT_DIR):
    if file.endswith(".txt"):
        # 파일 경로 설정
        file_path = os.path.join(TEXT_DIR, file)
        # 읽기 텍스트 내용 가져옴
        with open(file_path, "r", encoding="utf-8") as f:
            dog_text = f.read()

        try:
            print(f"처리 시작 : {file}")
            # LLM을 이용한 json 결과 값 생성
            result = create_json_with_llm(dog_text)
            # 저장 경로, 이름 변경
            output_path = os.path.join(JSON_DIR, f"{result['title']}.json")
            # 쓰기 json 파싱후 저장
            with open(output_path, "w", encoding="utf-8") as out:
                json.dump(result, out, ensure_ascii=False, indent=2)
            print(f"생성 완료: {output_path}")
        except Exception as e:
            print(f"처리 실패: {file}")
```

```python
import chromadb
from sentence_transformers import SentenceTransformer
import json

JSON_DIR = "./data/Json"

# Chromadb
client = chromadb.PersistentClient(path='./chroma_db')
# collection = client.create_collection(name = "dog_breed")
collection = client.get_collection(name= "dog_breed")

# embedding 모델 로드
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

def normalize_metadata(metadata: dict) -> dict:
    return {
        k: v if isinstance(v, (str, int, float, bool)) else str(v)
        for k, v in metadata.items()
    }

# json 파일 로드 후 입력
for file in os.listdir(JSON_DIR):
    if file.endswith(".json"):
        file_path = os.path.join(JSON_DIR, file)

        with open(file_path, "r", encoding="utf-8") as f:
            data = json.load(f)

        id = data["id"]
        title = data["title"]
        chunks = data["text_chunks"]
        metadata = normalize_metadata(data["metadata"])

        print(f"데이터 주입 시작 : {file}")

        # Chromadb에 데이터 입력, 임베딩
        for i, chunk in enumerate(chunks):
            embedding = embedding_model.encode([chunk]).tolist()[0]

            collection.add(
                ids=[f"{id}_chunk_{i}"],
                documents=[chunk],
                metadatas=[{
                    **metadata,
                    "title": title,
                    "chunk_index": i
                }],
                embeddings=[embedding]
            )

        print(f"데이터 주입 완료: {file}")


```

```python
import os
from openai import OpenAI
from dotenv import load_dotenv
from llm.call_llm import call_llm
from sentence_transformers import SentenceTransformer

load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

# 질문 입력
query = input("질문을 입력: ")

# 임베딩 모델 로드
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

# 질문 -> 임베딩 벡터 변환
query_vector = embedding_model.encode([query]).tolist()[0]

# Chroma 검색
results = collection.query(
    query_embeddings=[query_vector],
    n_results=3,
    include=["documents", "metadatas"]
)

# 검색된 텍스트 조합
retrieved_text = "\n\n".join(results["documents"][0])

print(f"유사 문구 : {retrieved_text}")

# 프롬프트 생성
system_prompt = "코기 견종 정보"
user_prompt = f"""
[질문]
{query}

[견종 정보]
{retrieved_text}
"""

# LLM
response = call_llm(system_prompt, user_prompt)

# 출력
print(f"\n답변: {response}")
```
